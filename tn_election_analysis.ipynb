```python
# ============================================================================
print("\n" + "="*80)
print("STEP 2: EXTRACTING WINNERS")
print("="*80)

def extract_winners(df, year_label):
    """Extract winner from each constituency"""
    winners_list = []
    
    for ac_num in sorted(df['ac_number'].unique()):
        ac_data = df[df['ac_number'] == ac_num].sort_values('votes', ascending=False)
        
        winner = ac_data.iloc[0]
        runner_up_votes = ac_data.iloc[1]['votes'] if len(ac_data) > 1 else 0
        margin = winner['votes'] - runner_up_votes
        
        winners_list.append({
            'ac_number': int(ac_num),
            'constituency': winner['constituency'],
            'winner': winner['candidate'],
            'party': winner['party'],
            'votes': int(winner['votes']),
            'runner_up_votes': int(runner_up_votes),
            'margin': int(margin),
            'reserved': winner['reserved'],
            'region': winner['region'],
            'year': year_label
        })
    
    return pd.DataFrame(winners_list)

winners_2021 = extract_winners(df_2021, 2021)
winners_2026 = extract_winners(df_2026, 2026)

print(f"\n✓ Extracted {len(winners_2021)} winners from 2021")
print(f"✓ Extracted {len(winners_2026)} winners from 2026")

# ============================================================================
print("\n" + "="*80)
print("PARTY DISTRIBUTION - THE DRAMATIC SHIFT")
print("="*80)

print("\n2021 Seats by Party (Top 10):")
print(winners_2021['party'].value_counts().head(10))

print("\n2026 Seats by Party (Top 10):")
print(winners_2026['party'].value_counts().head(10))

# ============================================================================
print("\n" + "="*80)
print("COMPARE THE SHIFT")
print("="*80)

comparison = pd.DataFrame({
    '2021_seats': winners_2021['party'].value_counts(),
    '2026_seats': winners_2026['party'].value_counts()
}).fillna(0).astype(int)

comparison['change'] = comparison['2026_seats'] - comparison['2021_seats']
comparison = comparison.sort_values('2026_seats', ascending=False)

print("\nTop 10 Parties - Seat Changes:")
print(comparison.head(10))

dmk_2021 = comparison.loc['DMK', '2021_seats'] if 'DMK' in comparison.index else 0
dmk_2026 = comparison.loc['DMK', '2026_seats'] if 'DMK' in comparison.index else 0
dmk_loss = dmk_2021 - dmk_2026

tvk_2026 = comparison.loc['TVK', '2026_seats'] if 'TVK' in comparison.index else 0

print(f"""
🔴 DMK (The Collapse):
   2021: {dmk_2021} seats
   2026: {dmk_2026} seats
   LOST: {dmk_loss} seats

🟢 TVK (The Disruption):
   2021: 0 seats
   2026: {tvk_2026} seats
   NEW PARTY: {tvk_2026} seats (46% of assembly!)
""")

# ============================================================================
print("\n" + "="*80)
print("SAVE WINNERS DATA")
print("="*80)

file_path = r"C:/Users/user/Desktop/SQL/codebasics challenge/input_files_for_participants_rpc/data"

winners_2021.to_csv(f'{file_path}/winners_2021.csv', index=False)
winners_2026.to_csv(f'{file_path}/winners_2026.csv', index=False)
comparison.to_csv(f'{file_path}/party_comparison.csv')

print("\n✓ Saved:")
print("  - winners_2021.csv")
print("  - winners_2026.csv")
print("  - party_comparison.csv")

print("\n" + "="*80)
print("✅ STEP 2 COMPLETE")
print("="*80)
```

```python
# ============================================================================
print("\n" + "="*80)
print("STEP 3: DETECTING FLIPS")
print("="*80)

# Merge 2021 and 2026 winners
flips = winners_2021.merge(
    winners_2026,
    on=['ac_number', 'constituency'],
    how='inner',
    suffixes=('_2021', '_2026')
)

print(f"\n✓ Merged data: {len(flips)} constituencies")

# Create flip indicator
flips['flipped'] = flips['party_2021'] != flips['party_2026']
flips['from_to'] = flips['party_2021'] + ' → ' + flips['party_2026']

total_seats = len(flips)
flipped_seats = flips['flipped'].sum()
unchanged_seats = total_seats - flipped_seats

print(f"""
FLIP STATISTICS:
  Total constituencies: {total_seats}
  Constituencies FLIPPED: {flipped_seats} ({flipped_seats/total_seats*100:.1f}%)
  Constituencies UNCHANGED: {unchanged_seats} ({unchanged_seats/total_seats*100:.1f}%)
""")

# ============================================================================
print("\n" + "="*80)
print("WHERE DID FLIPPED SEATS GO?")
print("="*80)

flipped_data = flips[flips['flipped']].copy()

print(f"\nFlipped seats - what party took over:")
print(flipped_data['party_2026'].value_counts())

print(f"\nFlipped seats - who lost:")
print(flipped_data['party_2021'].value_counts())

# ============================================================================
print("\n" + "="*80)
print("DMK'S COLLAPSE")
print("="*80)

dmk_flips = flips[(flips['party_2021'] == 'DMK') & (flips['party_2026'] != 'DMK')]
dmk_held = flips[(flips['party_2021'] == 'DMK') & (flips['party_2026'] == 'DMK')]

print(f"""
DMK STORY:
  Had in 2021: 133 seats
  Lost: {len(dmk_flips)} seats
  Held: {len(dmk_held)} seats
""")

print("\nWhere DMK's lost seats went:")
print(dmk_flips['party_2026'].value_counts())

# ============================================================================
print("\n" + "="*80)
print("TVK'S RISE")
print("="*80)

tvk_wins = flips[flips['party_2026'] == 'TVK']

print(f"\nTVK won {len(tvk_wins)} seats")
print("\nWhere TVK's seats came from:")
print(tvk_wins['party_2021'].value_counts())

# ============================================================================
print("\n" + "="*80)
print("REGIONAL FLIP ANALYSIS")
print("="*80)

flip_by_region = flips.groupby('region_2021').agg({
    'flipped': 'sum',
    'ac_number': 'count'
}).rename(columns={'flipped': 'flips', 'ac_number': 'total'})

flip_by_region['flip_pct'] = (flip_by_region['flips'] / flip_by_region['total'] * 100).round(1)
flip_by_region = flip_by_region.sort_values('flip_pct', ascending=False)

print("\nFlips by region:")
print(flip_by_region)

# ============================================================================
print("\n" + "="*80)
print("SAVE FLIP DATA")
print("="*80)

file_path = r"C:/Users/user/Desktop/SQL/codebasics challenge/input_files_for_participants_rpc/data"

flips.to_csv(f'{file_path}/flips_complete.csv', index=False)

flip_summary = {
    'total_seats': int(total_seats),
    'flipped_seats': int(flipped_seats),
    'flip_percentage': float(flipped_seats / total_seats * 100),
    'dmk_lost': int(len(dmk_flips)),
    'tvk_seats': int(len(tvk_wins)),
}

with open(f'{file_path}/flip_summary.json', 'w') as f:
    json.dump(flip_summary, f, indent=2)

print("\n✓ Saved:")
print("  - flips_complete.csv")
print("  - flip_summary.json")

print("\n" + "="*80)
print("✅ STEP 3 COMPLETE - FLIP STORY IS CLEAR")
print("="*80)
```

```python
# ============================================================================
print("\n" + "="*80)
print("STEP 4: VOTE SHARE ANALYSIS")
print("="*80)

# Calculate total votes by party (2021)
print("\n2021 VOTE SHARE BY PARTY:")
votes_2021 = df_2021[df_2021['party'] != 'NOTA'].groupby('party')['votes'].sum().sort_values(ascending=False)
total_votes_2021 = votes_2021.sum()
vote_share_2021 = (votes_2021 / total_votes_2021 * 100).round(2)

top_5_2021 = pd.DataFrame({
    'party': vote_share_2021.head(5).index,
    'vote_share_%': vote_share_2021.head(5).values,
    'total_votes': votes_2021.head(5).values
})
print(top_5_2021.to_string(index=False))

# Calculate total votes by party (2026)
print("\n\n2026 VOTE SHARE BY PARTY:")
votes_2026 = df_2026[df_2026['party'] != 'NOTA'].groupby('party')['votes'].sum().sort_values(ascending=False)
total_votes_2026 = votes_2026.sum()
vote_share_2026 = (votes_2026 / total_votes_2026 * 100).round(2)

top_5_2026 = pd.DataFrame({
    'party': vote_share_2026.head(5).index,
    'vote_share_%': vote_share_2026.head(5).values,
    'total_votes': votes_2026.head(5).values
})
print(top_5_2026.to_string(index=False))

# TVK specific analysis
tvk_votes_2026 = votes_2026.get('TVK', 0)
tvk_share_2026 = vote_share_2026.get('TVK', 0)

print(f"""
TVK ANALYSIS (2026):
  Total votes: {tvk_votes_2026:,}
  Vote share: {tvk_share_2026}%
  Seats won: 108
""")

# ============================================================================
print("\n" + "="*80)
print("VOTE SHARE BY REGION (2026)")
print("="*80)

regional_votes = []
for region in df_2026['region'].unique():
    region_data = df_2026[df_2026['region'] == region]
    tvk_votes_region = region_data[region_data['party'] == 'TVK']['votes'].sum()
    total_region_votes = region_data[region_data['party'] != 'NOTA']['votes'].sum()
    tvk_pct = (tvk_votes_region / total_region_votes * 100) if total_region_votes > 0 else 0
    
    regional_votes.append({
        'region': region,
        'tvk_vote_share_%': round(tvk_pct, 2),
        'tvk_votes': tvk_votes_region
    })

tvk_by_region = pd.DataFrame(regional_votes).sort_values('tvk_vote_share_%', ascending=False)
print(tvk_by_region.to_string(index=False))

# ============================================================================
print("\n" + "="*80)
print("SAVE VOTE SHARE DATA")
print("="*80)

file_path = r"C:/Users/user/Desktop/SQL/codebasics challenge/input_files_for_participants_rpc/data"

top_5_2026.to_csv(f'{file_path}/vote_share_2026.csv', index=False)
tvk_by_region.to_csv(f'{file_path}/tvk_regional.csv', index=False)

print("\n✓ Saved:")
print("  - vote_share_2026.csv")
print("  - tvk_regional.csv")

print("\n" + "="*80)
print("✅ STEP 4 COMPLETE - VOTE SHARE ANALYSIS DONE")
print("="*80)
```

```python
# ============================================================================
print("\n" + "="*80)
print("STEP 5: MARGIN OF VICTORY ANALYSIS")
print("="*80)

# Calculate margins
print("\n2021 MARGIN STATISTICS:")
margins_2021 = winners_2021['margin']
print(f"  Mean margin: {margins_2021.mean():.0f} votes")
print(f"  Median margin: {margins_2021.median():.0f} votes")
print(f"  Min margin: {margins_2021.min():.0f} votes")
print(f"  Max margin: {margins_2021.max():.0f} votes")

print("\n2026 MARGIN STATISTICS:")
margins_2026 = winners_2026['margin']
print(f"  Mean margin: {margins_2026.mean():.0f} votes")
print(f"  Median margin: {margins_2026.median():.0f} votes")
print(f"  Min margin: {margins_2026.min():.0f} votes")
print(f"  Max margin: {margins_2026.max():.0f} votes")

print("\n⚠️  Margins are SMALLER in 2026 - more competitive races!")

# Margin categories
print("\n\nMARGIN DISTRIBUTION (2026):")
winners_2026['margin_category'] = pd.cut(
    winners_2026['margin'],
    bins=[0, 5000, 10000, 20000, 50000, float('inf')],
    labels=['<5k', '5-10k', '10-20k', '20-50k', '>50k']
)

print(winners_2026['margin_category'].value_counts().sort_index())

# ============================================================================
print("\n" + "="*80)
print("RESERVED VS GENERAL SEATS MARGIN")
print("="*80)

margin_by_type = winners_2026.groupby('reserved')['margin'].agg(['mean', 'median', 'count'])
margin_by_type.columns = ['mean_margin', 'median_margin', 'count']
print(margin_by_type)

# ============================================================================
print("\n" + "="*80)
print("SAVE MARGIN DATA")
print("="*80)

file_path = r"C:/Users/user/Desktop/SQL/codebasics challenge/input_files_for_participants_rpc/data"

winners_2026.to_csv(f'{file_path}/winners_2026_with_margin.csv', index=False)

margin_summary = {
    'mean_2021': float(margins_2021.mean()),
    'mean_2026': float(margins_2026.mean()),
    'median_2021': float(margins_2021.median()),
    'median_2026': float(margins_2026.median()),
    'min_2026': int(margins_2026.min()),
    'max_2026': int(margins_2026.max())
}

with open(f'{file_path}/margin_summary.json', 'w') as f:
    json.dump(margin_summary, f, indent=2)

print("\n✓ Saved:")
print("  - winners_2026_with_margin.csv")
print("  - margin_summary.json")

print("\n" + "="*80)
print("✅ STEP 5 COMPLETE - MARGIN ANALYSIS DONE")
print("="*80)
```

```python
import plotly.graph_objects as go
import pandas as pd
import json

file_path = r"C:/Users/user/Desktop/SQL/codebasics challenge/input_files_for_participants_rpc/data"

# Load the outputs from steps 2-6 (already saved CSVs)
winners_2021 = pd.read_csv(f'{file_path}/winners_2021.csv')
winners_2026 = pd.read_csv(f'{file_path}/winners_2026_with_margin.csv')
df_2021 = pd.read_csv(f'{file_path}/tn_2021_results.csv')
df_2026 = pd.read_csv(f'{file_path}/tn_2026_results.csv')

# Rebuild flips correctly — merge on ac_number ONLY
flips = winners_2021.merge(winners_2026, on='ac_number', suffixes=('_2021','_2026'))
flips['flipped'] = flips['party_2021'] != flips['party_2026']
flipped_data = flips[flips['flipped']]

# ── CHART 1: Seats comparison ─────────────────────────────────────────────
seats_2021 = winners_2021['party'].value_counts()
seats_2026 = winners_2026['party'].value_counts()
parties = ['TVK','DMK','AIADMK','INC','PMK','CPI','VCK','CPI(M)']
y2021 = [int(seats_2021.get(p,0)) for p in parties]
y2026 = [int(seats_2026.get(p,0)) for p in parties]

fig1 = go.Figure(data=[
    go.Bar(name='2021', x=parties, y=y2021, marker_color='#1f77b4'),
    go.Bar(name='2026', x=parties, y=y2026, marker_color='#ff7f0e')
])
fig1.update_layout(title='Assembly Seats Won - 2021 vs 2026',
    xaxis_title='Party', yaxis_title='Seats', barmode='group', height=500)
fig1.write_html(f'{file_path}/chart_1_seats_comparison.html')
print(f"Chart 1 — 2021: {y2021}, 2026: {y2026}")

# ── CHART 2: Sankey ───────────────────────────────────────────────────────
# DMK losses
dmk_lost = flipped_data[flipped_data['party_2021']=='DMK']['party_2026'].value_counts()
aiadmk_lost = flipped_data[flipped_data['party_2021']=='AIADMK']['party_2026'].value_counts()
inc_lost = flipped_data[flipped_data['party_2021']=='INC']['party_2026'].value_counts()

dmk_to_tvk = int(dmk_lost.get('TVK',0))
dmk_to_aiadmk = int(dmk_lost.get('AIADMK',0))
dmk_to_others = int(dmk_lost.sum()) - dmk_to_tvk - dmk_to_aiadmk
aiadmk_to_tvk = int(aiadmk_lost.get('TVK',0))
aiadmk_to_dmk = int(aiadmk_lost.get('DMK',0))
aiadmk_to_others = int(aiadmk_lost.sum()) - aiadmk_to_tvk - aiadmk_to_dmk
inc_to_tvk = int(inc_lost.get('TVK',0))
inc_to_others = int(inc_lost.sum()) - inc_to_tvk

print(f"Chart 2 — DMK→TVK:{dmk_to_tvk}, DMK→AIADMK:{dmk_to_aiadmk}, DMK→others:{dmk_to_others}")
print(f"          AIADMK→TVK:{aiadmk_to_tvk}, AIADMK→DMK:{aiadmk_to_dmk}, AIADMK→others:{aiadmk_to_others}")
print(f"          INC→TVK:{inc_to_tvk}, INC→others:{inc_to_others}")

# nodes: 0=DMK21, 1=AIADMK21, 2=INC21, 3=TVK26, 4=AIADMK26, 5=DMK26, 6=Others26
fig2 = go.Figure(data=[go.Sankey(
    node=dict(pad=15, thickness=20,
        label=['DMK (2021)','AIADMK (2021)','INC (2021)','TVK (2026)','AIADMK (2026)','DMK (2026)','Others (2026)'],
        color=['#d62728','#9467bd','#2ca02c','#ff7f0e','#9467bd','#d62728','#8c564b']),
    link=dict(
        source=[0,0,0, 1,1,1, 2,2],
        target=[3,4,6, 3,5,6, 3,6],
        value=[dmk_to_tvk, dmk_to_aiadmk, dmk_to_others,
               aiadmk_to_tvk, aiadmk_to_dmk, aiadmk_to_others,
               inc_to_tvk, inc_to_others])
)])
fig2.update_layout(title='Where Did Seats Flow? 2021→2026', height=600)
fig2.write_html(f'{file_path}/chart_2_sankey_flow.html')

# ── CHART 3: Regional analysis ────────────────────────────────────────────
# Flip rates — correct merge on ac_number only
flip_by_region = flips.groupby('region_2021').agg(
    flips=('flipped','sum'), total=('ac_number','count'))
flip_by_region['flip_pct'] = (flip_by_region['flips']/flip_by_region['total']*100).round(1)
flip_by_region = flip_by_region.sort_values('flip_pct', ascending=False)

# TVK vote share by region — from raw 2026 results
tvk_share = df_2026.groupby('region').apply(
    lambda x: x[x['party']=='TVK']['votes'].sum() /
              x[x['party']!='NOTA']['votes'].sum() * 100
).round(1)

regions_ordered = flip_by_region.index.tolist()
flip_rates = flip_by_region['flip_pct'].tolist()
tvk_pcts = [float(tvk_share.get(r, 0)) for r in regions_ordered]

print(f"\nChart 3 — Regions: {regions_ordered}")
print(f"          Flip rates: {flip_rates}")
print(f"          TVK vote%: {tvk_pcts}")

fig3 = go.Figure(data=[
    go.Bar(name='Flip Rate (%)', x=regions_ordered, y=flip_rates,
           marker_color='#d62728', yaxis='y'),
    go.Scatter(name='TVK Vote Share (%)', x=regions_ordered, y=tvk_pcts,
               marker=dict(color='#ff7f0e', size=10),
               line=dict(color='#ff7f0e', width=3), yaxis='y2')
])
fig3.update_layout(title='Regional Performance: Flip Rate & TVK Vote Share',
    xaxis_title='Region',
    yaxis=dict(title='Flip Rate (%)', side='left', range=[0,100]),
    yaxis2=dict(title='TVK Vote Share (%)', overlaying='y', side='right', range=[0,60]),
    height=500, hovermode='x unified')
fig3.write_html(f'{file_path}/chart_3_regional_analysis.html')

# ── CHART 4: Margin distribution ──────────────────────────────────────────
bins   = [0, 5000, 10000, 20000, 30000, 50000, 999999]
labels = ['0–5K','5–10K','10–20K','20–30K','30–50K','50K+']
margin_dist = pd.cut(winners_2026['margin'], bins=bins, labels=labels).value_counts().sort_index()
print(f"\nChart 4 — Margin bins: {margin_dist.tolist()}")

fig4 = go.Figure(data=[go.Bar(
    x=labels, y=margin_dist.tolist(),
    marker_color=['#d62728','#ff7f0e','#2ca02c','#1f77b4','#9467bd','#8c564b']
)])
fig4.update_layout(title='Victory Margin Distribution (2026)',
    xaxis_title='Victory Margin', yaxis_title='Number of Constituencies', height=500)
fig4.write_html(f'{file_path}/chart_4_margin_distribution.html')

# ── CHART 5: Vote share pies ──────────────────────────────────────────────
votes_2021 = df_2021[df_2021['party']!='NOTA'].groupby('party')['votes'].sum()
total_21 = votes_2021.sum()
vs21 = (votes_2021/total_21*100).round(1).sort_values(ascending=False)

votes_2026 = df_2026[df_2026['party']!='NOTA'].groupby('party')['votes'].sum()
total_26 = votes_2026.sum()
vs26 = (votes_2026/total_26*100).round(1).sort_values(ascending=False)

# Top 5 + Others for each year
top5_21 = vs21.head(5)
others_21 = round(100 - top5_21.sum(), 1)
top5_26 = vs26.head(5)
others_26 = round(100 - top5_26.sum(), 1)

labels_21 = top5_21.index.tolist() + ['Others']
vals_21   = top5_21.tolist() + [others_21]
labels_26 = top5_26.index.tolist() + ['Others']
vals_26   = top5_26.tolist() + [others_26]

print(f"\nChart 5 — 2021: {list(zip(labels_21, vals_21))}")
print(f"          2026: {list(zip(labels_26, vals_26))}")

fig5 = go.Figure()
fig5.add_trace(go.Pie(labels=labels_21, values=vals_21, name='2021',
    hole=0.4, domain=dict(x=[0,0.45]),
    marker_colors=['#d62728','#9467bd','#7f7f7f','#2ca02c','#1f77b4','#c7c7c7']))
fig5.add_trace(go.Pie(labels=labels_26, values=vals_26, name='2026',
    hole=0.4, domain=dict(x=[0.55,1]),
    marker_colors=['#ff7f0e','#d62728','#9467bd','#7f7f7f','#2ca02c','#c7c7c7']))
fig5.update_layout(title='Vote Share: 2021 vs 2026', height=500,
    annotations=[dict(text='2021',x=0.225,y=0.5,font_size=16,showarrow=False),
                 dict(text='2026',x=0.775,y=0.5,font_size=16,showarrow=False)])
fig5.write_html(f'{file_path}/chart_5_vote_share.html')

# ── CHART 6: Reserved seats ───────────────────────────────────────────────
tvk_gen = int(((winners_2026['reserved']=='GEN') & (winners_2026['party']=='TVK')).sum())
tvk_sc  = int(((winners_2026['reserved']=='SC')  & (winners_2026['party']=='TVK')).sum())
tvk_st  = int(((winners_2026['reserved']=='ST')  & (winners_2026['party']=='TVK')).sum())
pct_gen = round(tvk_gen/188*100, 1)
pct_sc  = round(tvk_sc/44*100, 1)
pct_st  = round(tvk_st/2*100, 1)

print(f"\nChart 6 — TVK wins: GEN={tvk_gen}({pct_gen}%), SC={tvk_sc}({pct_sc}%), ST={tvk_st}({pct_st}%)")

fig6 = go.Figure(data=[
    go.Bar(name='TVK Wins', x=['General (188)','SC Reserved (44)','ST Reserved (2)'],
           y=[tvk_gen, tvk_sc, tvk_st],
           marker_color=['#ff7f0e','#2ca02c','#1f77b4'], yaxis='y'),
    go.Scatter(name='TVK Win %',
               x=['General (188)','SC Reserved (44)','ST Reserved (2)'],
               y=[pct_gen, pct_sc, pct_st],
               marker=dict(color='#d62728', size=12),
               line=dict(color='#d62728', width=3), yaxis='y2')
])
fig6.update_layout(title='TVK Performance: General vs Reserved Seats',
    yaxis=dict(title='Number of Wins', side='left'),
    yaxis2=dict(title='Win Percentage (%)', overlaying='y', side='right', range=[0,70]),
    height=500, hovermode='x unified')
fig6.write_html(f'{file_path}/chart_6_reserved_seats.html')

print("\n✅ All 6 charts rebuilt with verified data.")
```
